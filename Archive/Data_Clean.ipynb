```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# Store filepath in a variable
Poverty_Data = "Data/Poverty_Data.csv"
Poverty_Education_Data = "Data/Poverty_Education_Data.csv"
Income_Data = "Data/Income.csv"
Education_Income_Data = "Data/ACS_17_5YR_S1501_Education.csv"
Married_Data = "Data/Married_By_Zip.csv"
```

```python
Poverty_Data_df = pd.read_csv(Poverty_Data, encoding="ISO-8859-1", skiprows=[1], low_memory=False)
Poverty_Education_Data_df = pd.read_csv(Poverty_Education_Data, encoding="ISO-8859-1", skiprows=[1], low_memory=False)
Income_Data_df = pd.read_csv(Income_Data, encoding="ISO-8859-1", skiprows=[1], low_memory=False)
Education_Income_Data_df = pd.read_csv(Education_Income_Data, encoding="ISO-8859-1", skiprows=[1], low_memory=False)
Married_Data_df = pd.read_csv(Married_Data, encoding="ISO-8859-1", low_memory=False)
```

```python
Married_Data_df.head()
```

```python
Poverty_Data_df['GEO.display-label'] = Poverty_Data_df['GEO.display-label'].astype(str).str[-5:]
Poverty_Education_Data_df['GEO.display-label'] = Poverty_Education_Data_df['GEO.display-label'].astype(str).str[-5:]
Income_Data_df['GEO.display-label'] = Income_Data_df['GEO.display-label'].astype(str).str[-5:]
Education_Income_Data_df['GEO.display-label'] = Education_Income_Data_df['GEO.display-label'].astype(str).str[-5:]
Married_Data_df['Zip Code'] = Married_Data_df['Zip Code'].astype(str).str[-5:]
```

```python
Married_Data_df.head()
```

```python
del Poverty_Data_df['GEO.id']
del Poverty_Data_df['GEO.id2']
Poverty_Data_df.rename(columns={'GEO.display-label':'Zip Code'}, inplace=True)

del Poverty_Education_Data_df['GEO.id']
del Poverty_Education_Data_df['GEO.id2']
Poverty_Education_Data_df.rename(columns={'GEO.display-label':'Zip Code'}, inplace=True)

del Income_Data_df['GEO.id']
del Income_Data_df['GEO.id2']
Income_Data_df.rename(columns={'GEO.display-label':'Zip Code'}, inplace=True)

del Education_Income_Data_df['GEO.id']
del Education_Income_Data_df['GEO.id2']
Education_Income_Data_df.rename(columns={'GEO.display-label':'Zip Code'}, inplace=True)
```

```python
Income_Data_df.rename(columns={'HC03_EST_VC02':'Mean income (dollars); Estimate; All households', 'HC03_MOE_VC02':'Mean income (dollars); Margin of Error; All households'}, inplace=True)
Education_Income_Data_df.rename(columns={'HC01_EST_VC81':'MEDIAN EARNINGS IN THE PAST 12 MONTHS (IN 2017 INFLATION-ADJUSTED DOLLARS) - Population 25 years and over with earnings - Less than high school graduate',\
                                 'HC01_EST_VC82' : 'MEDIAN EARNINGS IN THE PAST 12 MONTHS (IN 2017 INFLATION-ADJUSTED DOLLARS) - Population 25 years and over with earnings - High school graduate (includes equivalency)',\
                                 'HC01_EST_VC83' : 'MEDIAN EARNINGS IN THE PAST 12 MONTHS (IN 2017 INFLATION-ADJUSTED DOLLARS) - Population 25 years and over with earnings - Some college or associate\'s degree',\
                                 'HC01_EST_VC84' : 'MEDIAN EARNINGS IN THE PAST 12 MONTHS (IN 2017 INFLATION-ADJUSTED DOLLARS) - Population 25 years and over with earnings - Bachelor\'s degree',\
                                 'HC01_EST_VC85' : 'MEDIAN EARNINGS IN THE PAST 12 MONTHS (IN 2017 INFLATION-ADJUSTED DOLLARS) - Population 25 years and over with earnings - Graduate or professional degree',\
                                 'HC02_EST_VC09' : 'Percent; Estimate; Population 25 years and over - Less than 9th grade',\
                                 'HC02_EST_VC10' : 'Percent; Estimate; Population 25 years and over - 9th to 12th grade, no diploma',\
                                 'HC01_EST_VC11' : 'Total; Estimate; Population 25 years and over - High school graduate (includes equivalency)',\
                                 'HC02_EST_VC12' : 'Percent; Estimate; Population 25 years and over - Some college, no degree',\
                                 'HC02_EST_VC13' : 'Percent; Estimate; Population 25 years and over - Associate\'s degree',\
                                 'HC02_EST_VC14' : 'Percent; Estimate; Population 25 years and over - Bachelor\'s degree',\
                                 'HC02_EST_VC15' : 'Percent; Estimate; Population 25 years and over - Graduate or professional degree'\
                                 ''}, inplace=True)
```

```python
Income_Data_clean_df = Income_Data_df[['Zip Code','Mean income (dollars); Estimate; All households', 'Mean income (dollars); Margin of Error; All households']]
Education_Income_Data_clean_df = Education_Income_Data_df[['Zip Code',\
                                           'MEDIAN EARNINGS IN THE PAST 12 MONTHS (IN 2017 INFLATION-ADJUSTED DOLLARS) - Population 25 years and over with earnings - Less than high school graduate',\
                                           'MEDIAN EARNINGS IN THE PAST 12 MONTHS (IN 2017 INFLATION-ADJUSTED DOLLARS) - Population 25 years and over with earnings - High school graduate (includes equivalency)',\
                                           'MEDIAN EARNINGS IN THE PAST 12 MONTHS (IN 2017 INFLATION-ADJUSTED DOLLARS) - Population 25 years and over with earnings - Some college or associate\'s degree',\
                                           'MEDIAN EARNINGS IN THE PAST 12 MONTHS (IN 2017 INFLATION-ADJUSTED DOLLARS) - Population 25 years and over with earnings - Bachelor\'s degree',\
                                           'MEDIAN EARNINGS IN THE PAST 12 MONTHS (IN 2017 INFLATION-ADJUSTED DOLLARS) - Population 25 years and over with earnings - Graduate or professional degree']]
Education_Attained_Data_clean_df = Education_Income_Data_df[['Zip Code',\
                                           'Percent; Estimate; Population 25 years and over - Less than 9th grade',\
                                           'Percent; Estimate; Population 25 years and over - 9th to 12th grade, no diploma',\
                                           'Percent; Estimate; Population 25 years and over - Some college, no degree',\
                                           'Percent; Estimate; Population 25 years and over - Associate\'s degree',\
                                           'Percent; Estimate; Population 25 years and over - Bachelor\'s degree',\
                                           'Percent; Estimate; Population 25 years and over - Graduate or professional degree']]
```

```python
Education_Attained_Data_clean_df.head()
```

```python
combined_data_df = pd.merge(Income_Data_clean_df, Education_Income_Data_clean_df, how="left", on="Zip Code")
combined_data_df = pd.merge(combined_data_df, Education_Attained_Data_clean_df, how="left", on="Zip Code")
combined_data_df = pd.merge(combined_data_df, Married_Data_df, how="left", on="Zip Code")
```
